```
import matplotlib.pyplot as plt
import numpy as np
genders = ['Male', 'Female', 'Non-binary']
data = np.random.choice(genders, size=1000, p=[0.45, 0.50, 0.05])
counts = np.unique(data, return_counts=True)
categories = counts[0]
frequencies = counts[1]
plt.figure(figsize=(8, 6))
plt.bar(categories, frequencies, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribution of Sample Genders')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
